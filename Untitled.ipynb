```python
error_allowed = 0.005



w, h = 10, 10;

ind = [[0 for x in range(w)] for y in range(h)] 

sep = [[0 for x in range(w)] for y in range(h)] 

suf = [[0 for x in range(w)] for y in range(h)] 


i = 6

print('\n \n evaluating different criterias with one threshold : Independence,  Seperation and sufficiency for the threshold',thr_x[i])


print('\n Independency : ')


if(abs((TN[0][i]+FN[0][i])/(TN[0][i]+FN[0][i]+TP[0][i]+FP[0][i])-(TN[3][i]+FN[3][i])/(TN[3][i]+FN[3][i]+TP[3][i]+FP[3][i])) <= error_allowed ):
    print('independence fulfilled')
else:
    print('independence not fulfilled')
if(abs(TP[0][i]/(TP[0][i]+FP[0][i])-TP[3][i]/(TP[3][i]+FP[3][i])) <= error_allowed ):
    print('seperation fulfilled')
else:
    print('seperation not fulfilled')

if(abs(TP[0][i]/(TP[0][i]+FN[0][i])-TP[3][i]/(TP[3][i]+FN[3][i])) <= error_allowed ) :
    print('sufficiency fulfilled')
else:
    print('sufficiency not fulfilled')

    
    
for i in range(10):
    for j in range(10):
        if(abs((TP[0][i]+FP[0][i])/(TN[0][i]+FN[0][i]+TP[0][i]+FP[0][i])-(TP[3][j]+FP[3][j])/(TN[3][j]+FN[3][j]+TP[3][j]+FP[3][j])) <= 0.01 ):
            ind[i][j]= 1
        else:
            ind[i][j]= 0
        if(abs(TP[0][i]/(TP[0][i]+FP[0][i])-TP[3][j]/(TP[3][j]+FP[3][j])) <= 0.01 ):
            sep[i][j]= 1
        else:
            sep[i][j]= 0

        if(abs(TP[0][i]/(TP[0][i]+FN[0][i])-TP[3][j]/(TP[3][j]+FN[3][j])) <= 0.01 ) :
            suf[i][j] = 1
        else:
            suf[i][j] = 0

print('\n', ind)
print('\n', sep)
print('\n', suf)



print('Independency comparison for different thresholds :')


cm = sns.light_palette("blue", as_cmap=True)
x=pd.DataFrame(ind)
x=x.style.background_gradient(cmap=cm)
display(x)


print('Seperation comparison for different thresholds :')

cm = sns.light_palette("blue", as_cmap=True)
x=pd.DataFrame(sep)
x=x.style.background_gradient(cmap=cm)
display(x)



print('sufficiency comparison for different thresholds :')

cm = sns.light_palette("blue", as_cmap=True)
x=pd.DataFrame(suf)
x=x.style.background_gradient(cmap=cm)
display(x)



crit = [[ind[i][j] + suf[i][j] + sep[i][j]  for j in range(h)] for i in range(w)]



print('overall comparison for different thresholds :')

cm = sns.light_palette("green", as_cmap=True)
x=pd.DataFrame(crit)
x=x.style.background_gradient(cmap=cm)
display(x)








'''
for i in range(10):
    print('\n \n evaluating : Independence,  Seperation and sufficiency for the threshold',thr_x[i])
    
    
    if(abs((TP[0][i]+FP[0][i])/(TN[0][i]+FN[0][i]+TP[0][i]+FP[0][i])-(TP[3][i]+FP[3][i])/(TN[3][i]+FN[3][i]+TP[3][i]+FP[3][i])) <= 0.01 ):
        print('independence fulfilled')
        ind[i] = 1
    else:
        print('independence not fulfilled')
        ind[i] = 0
    if(abs(TP[0][i]/(TP[0][i]+FP[0][i])-TP[3][i]/(TP[3][i]+FP[3][i])) <= 0.01 ):
        print('seperation fulfilled')
        sep[i] = 1
    else:
        print('seperation not fulfilled')
        sep[i] = 0

    if(abs(TP[0][i]/(TP[0][i]+FN[0][i])-TP[3][i]/(TP[3][i]+FN[3][i])) <= 0.01 ) :
        print('sufficiency fulfilled')
        suf[i] = 1
    else:
        print('sufficiency not fulfilled')
        suf[i] = 0
        
    '''

print()
```
