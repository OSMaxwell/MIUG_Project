```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
data = pd.read_csv('compas_rec.csv')
data.head(50)
```

```python

data['dummy'] = np.ones(shape = data.shape[0])
cat_cols=['Race','Sex','Age','Days in Jail']
for col in cat_cols:
    print(col)
    counts = data[['dummy', 'Recidivism', col]].groupby(['Recidivism', col], as_index = False).count()
    temp = counts[counts['Recidivism'] == 0][[col, 'dummy']]
    _ = plt.figure(figsize = (10,4))
    plt.subplot(1, 2, 1)
    temp = counts[counts['Recidivism'] == 0][[col, 'dummy']]
    plt.bar(temp[col], temp['dummy'])
    plt.xticks(rotation=90)
    plt.title('Counts for ' + col + '\n Recidivist')
    plt.ylabel('count')
    plt.subplot(1, 2, 2)
    temp = counts[counts['Recidivism'] == 1][[col, 'dummy']]
    plt.bar(temp[col], temp['dummy'])
    plt.xticks(rotation=90)
    plt.title('Counts for ' + col + '\n not Recidivist')
    plt.ylabel('count')
    plt.show()
```

```python
def plot_box(data, cols):
    for col in cols:
        sns.set_style("whitegrid")
        sns.boxplot('Recidivism', col, data=data)
        plt.xlabel('Recidivism') 
        plt.ylabel(col)
        plt.show()

cols=['Age','Previous Convictions','Days in Jail']
plot_box(data, cols)
```
